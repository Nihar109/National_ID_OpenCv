```python
## Regular Expression TO frame them in Adhar Name,DOB,Gender,UID
```

```python
import re
names= re.findall(r'[A-Z][A-Z]*',text)
print(names)
```

```python
UID=re.findall("\w{4}\s\w{4}\s\w{4}",text)
print(UID)
```

```python
Name=re.findall("\w{5}\s\w{6}\s\w{5}",text)
print(Name)
```

```python
Firstname=Name[0][0:12]
Lastname=Name[0][13:]
print(Firstname)
print(Lastname)
```

```python
Gender=None
if 'Male' in text:
  Gender='Male'
else:
  Gender='Female'
print(Gender)
```

```python
Date= re.findall(r'\d{2}/\d{2}/\d{4}',text)
print(Date)
```

```python
DOB=Date[1]
DOB
```
